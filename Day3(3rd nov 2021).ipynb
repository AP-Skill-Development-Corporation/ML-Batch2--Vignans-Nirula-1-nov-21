### Machine Learning
- A part of Aritificial Intelligence (it is also a part of Data Science).
- OR A subset of AI which used to make predictions based on previous data .
- Father of ML is ****
* **Types of ML**
    1. Supervised ML
        - we can predict the o/p data based on given i/p data
        - works on labeled data means(posses both i/p and o/p dataset)
    2. Un Supervised ML and
        - which have only i/p data but not have o/p data
            - works on unlabeled data
            - (self learning)
    3. Reinforcement ML
        - seeing the final o/p in environment
        - Environmental ML
        - still in process(trails)

##### Supervised Machine Learning
- Regression & Classification
    - **Regression**
        - Continuous data:
        - Ex:Height,weight,sales,price etc
    - **Classifiction**
        - Categorical data
        - Ex:Gender,Colors etc
        - **2types**
           - Binary Classification and 
           - Multi Classification

#### Unsupervised ML
- 2types
    - Clusters and 
        - Groups
    - Dimensionality Reduction(DMR)
        - Reduces the no.of coloumns/dimensions

#### Reinforcement ML
- Environmental Machine Learning
- A model has build to pass instructions to a rabit which should go across the path to reach destination(puzzle)

#### Linear Regression
- It is a linear approach to find the relation b/w both i/p and o/p data
- 2types<br>
    1. simple linear reg &
        - finding the relation b/w one i/p and one o/p data
        * formula*
            - y=mx+c 
            - where y =o/p var,m=slope c=intercept and x= i/p var
    2. multiple linear reg
        - finding the relation among 2 or more i/ps(indepedent var) and 1 o/p data(dependent var)
        - *formula:*
            - y=m1x1+mx2+m3x3...c
